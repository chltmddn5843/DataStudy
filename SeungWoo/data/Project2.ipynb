## 카드이용이 많은 지역과 업종 알아보기

```python
# 업종별, 지역별 카드 이용 현황 출력하기
jeju_17_18_df['업종명']
groupby_sector = jeju_17_18_df.groupby('업종명').sum(numeric_only=True)
```

```python
# 이용금액 상위 10위 업종 알아보기
groupby_sector.sort_values(by='이용금액',ascending=False).head(10)
# 올림차수로 하면 작은수부터이므로 False로 하고 head(10)
```

```python
groupby_sector.sort_values(by='이용자수', ascending=False).head(10)
#이용금액과 이용자수로 조건에 따라 차이가 있음
```

```python
groupby_sector['인당이용금액'] = groupby_sector['이용금액'] / groupby_sector['이용자수']
# 1인당 이용금액
```

```python
groupby_sector.sort_values(by='이용자수', ascending=False).head(10)
```

```python
#인당 이용금액을 기준으로 상위 10개 출력
groupby_sector.sort_values(by='인당이용금액', ascending=False).head(10)
# 갬블링,베팅업 / 버스는 3명 5명 뿐이네 이용금액이 작을 수밖에 없다
```

```python
jeju_17_18_df = jeju_17_18_df[jeju_17_18_df['업종명'] != '버스 운송업']
jeju_17_18_df = jeju_17_18_df[jeju_17_18_df['업종명'] != '기타 갬블링 및 베팅업']
# 버스 운송업은 제외해보자
```

```python
groupby_sector.sort_values(by='인당이용금액', ascending=False).head(10)
```

```python
jeju_reg_df = jeju_17_18_df[jeju_17_18_df['업종명'] != '버스 운송업']
```

```python
# 업종별 합계 및 인당 이용금액 계산
groupby_sector = jeju_reg_df.groupby('업종명').sum(numeric_only=True)
groupby_sector['인당이용금액'] = groupby_sector['이용금액'] / groupby_sector['이용자수']

# 인당 이용금액 상위 10개 업종
groupby_sector.sort_values(by='인당이용금액', ascending=False).head(10)
```
